#Q1) a) Calculate pmf of Histogram store it in 'hist_pmf'. [1 Marks]
#b) Calculate cdf of that pmf and store it in 'hist_cdf'. [1 Marks]  

```python
# Function to calculate PMF from bin and array
total_pixels = img.size
hist_pmf = hist / total_pixels
# Print the PMF array
print("PMF Array:")
print(hist_pmf)
print(f"\n")

# Function to calculate CDF from PMF
hist_cdf = np.cumsum(hist_pmf)
# Print the CDF array
print("CDF Array:")
print(hist_cdf)
print(f"\n")

# Plot the image and histogram
plt.figure(figsize=(10, 5))

# Plot the image
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title('Original Image')


# Plot the PMF
plt.subplot(1, 3, 2)
plt.plot(hist_pmf, color='blue', label='PMF')
plt.title('PMF')
plt.xlabel('Pixel Value')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)

# Plot the CDF
plt.subplot(1, 3, 3)
plt.plot(hist_cdf, color='red', label='CDF')
plt.title('CDF')
plt.xlabel('Pixel Value')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.grid(True)
```

#Q2) Calculate $T(r)$ [3 Marks]

```python
# Define the transformation function t(r)
t_r = (hist_cdf /hist_cdf[-1]) * 255

```

#Q3) Apply $T(r)$ on input image. [2 Marks]
#Store output image in 'out_img1'

```python
# Apply the transformation function t(r) to the original image
out_img1 = np.interp(img.flatten(), range(256), t_r)
out_img1 = out_img1.reshape(img.shape)\

out_img1 = np.uint8(out_img1)
```

#Q4) Implement Histogram matching by CDF based transformation function. [3 Marks]

```python
def histogram_matching(img, ref):
    # Calculating histograms and CDFs for the input and reference images
    hist_img, bins_img = np.histogram(img.flatten(), 256, [0, 256])
    cdf_img = np.cumsum(hist_img) / np.sum(hist_img)

    hist_ref, bins_ref = np.histogram(ref.flatten(), 256, [0, 256])
    cdf_ref = np.cumsum(hist_ref) / np.sum(hist_ref)

    # Mapping each intensity value in the input image to the corresponding value in the reference CDF
    matched_img = np.interp(img.flatten(), bins_img[:-1], np.interp(bins_img[:-1], bins_ref[:-1], cdf_ref))
    matched_img = matched_img.reshape(img.shape)
    matched_img = np.uint8(255 * matched_img / matched_img.max())

    return matched_img, hist_img, bins_img, hist_ref, bins_ref

# Perform histogram matching
matched_img, hist_inp, bins_inp, hist_ref, bins_ref = histogram_matching(inp, ref)
```
